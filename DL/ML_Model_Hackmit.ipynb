```python
import pandas as pd

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('new_TRAIN_final.csv')
df.head()
```

```python
properties = list(df.columns.values)
properties.remove('Activity')
print(properties)
X = df[properties]
y = df['Activity']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(11,)),
    keras.layers.Dense(121, activation=tf.nn.relu, name="layer_1"),
    keras.layers.Dense(121, activation=tf.nn.relu, name="layer_2"),
    keras.layers.Dense(1, activation=tf.nn.sigmoid, name="layer_3"),
])
```

```python
# embed_dim = 128
# lstm_out = 196

# model = tf.keras.Sequential()
# model.add(tf.keras.layers.Embedding(max_fatures, embed_dim, input_length=X.shape[1]))
# model.add(tf.keras.layers.SpatialDropout1D(0.4))
# model.add(tf.keras.layers.LSTM(lstm_out, dropout=0.3,
#                recurrent_dropout=0.2, return_sequences=True))
# model.add(tf.keras.layers.LSTM(128, recurrent_dropout=0.2))
# model.add(tf.keras.layers.Dense(3, activation='softmax'))

# model.compile(loss='categorical_crossentropy',
#               optimizer='adam', metrics=['accuracy'])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=25, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
model.save('FINAL_MODEL.h5')
```
